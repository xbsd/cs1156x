```python
def transform(x0, x1, x2, phi):
    return np.matrix ([x0, x1, x2, x1**2, x2**2, x1*x2, np.abs(x1 - x2), np.abs(x1 + x2)])[:phi+1]

def getW(Z, y_value):
    return np.dot(np.dot(np.linalg.inv(np.dot(Z, Z.T)), Z), y)

def getError(x0,x1,x2,y, phi):
    Z = transform(x0, x1, x2, phi)
    W = np.dot(np.dot(np.linalg.inv(np.dot(Z, Z.T)), Z), y)
    result = np.sign(W * Z)
    misclassified = np.sum((result != y)*1)
    return list([misclassified/np.float(Z.shape[1]), W])

```

```python
# Clean Restart
# Question 8

def genPtsAndPlot(n, plot='Y'):


    # Random n points
    xvals = np.random.uniform(-1,1,n)
    yvals = np.random.uniform(-1,1,n)

    # Pick 2 Random Points to draw a line
    xrnd = np.random.uniform(-1,1,2)
    yrnd = np.random.uniform(-1,1,2)

    output = np.zeros(n)
    grad_line = (yrnd[1] - yrnd[0])/(xrnd[1] - xrnd[0])
    intercept = yrnd[0] - grad_line*xrnd[0]

    for i in range(n):
        x = xvals[i]
        y = yvals[i]
        line_x_val = (y - intercept)/grad_line
        output[i] = np.sign(x - line_x_val)

    # Note
    # output variable stores the +/- 1 indication
    #output[output==0]=-1

    # Draw the demarcating line

    # At y = 1, x = ...
    xrnd1 = (1-intercept)/grad_line

    # At y = -1
    xrnd_minus1 = (-1-intercept)/grad_line

    if plot == 'Y':
        # Plot Complete Figure
        plt.figure()
        plt.scatter(xvals, yvals, alpha=0.5, s = 50, c = output)
        plt.plot([xrnd1, xrnd_minus1], [1,-1], color='green')
        plt.show()

    return ({'xvals':xvals, 'yvals':yvals, 'output':output, 'grad_line':grad_line, \
             'xrnd1':xrnd1, 'xrnd_minus1':xrnd_minus1, 'intercept':intercept})




def getEout(n_out, new_xvals, new_yvals, orig_intercept, orig_grad_line, w, plot_flag = 'Y'):
    output_out = np.zeros(n_out)
    constant_out = np.ones(n_out)

    new_out = genPtsAndPlot(n_out,'N')
    new_xvals = new_out['xvals']
    new_yvals = new_out['yvals']

    for i in range(n_out):
        x = new_xvals[i]
        y = new_yvals[i]
        line_x_val = (y - orig_intercept)/orig_grad_line
        output_out[i] = np.sign(x - line_x_val)


    # Note
    # output variable stores the +/- 1 indication
    output_out[output_out==0]=-1


    # At y = 1, x = ...
    orig_xrnd1 = (1-orig_intercept)/orig_grad_line

    # At y = -1
    orig_xrnd_minus1 = (-1-orig_intercept)/orig_grad_line


    if plot_flag == 'Y':
        # Plot Complete Figure
        plt.figure()
        plt.scatter(new_xvals, new_yvals, alpha=0.5, s = 50, c = output_out)
        plt.plot([orig_xrnd1, orig_xrnd_minus1], [1,-1], color='green')
        plt.show()


    # Calculate Eout

    x_out = pd.Series(np.linspace(0,n_out-1,n_out))
    x_out_vector = np.matrix([constant_out, new_xvals, new_yvals])
    x_out_vector = x_out_vector.T

    E_out = (1./n_out) * np.sum(x_out.apply(lambda p: np.log(1 + e ** ((-output_out[p] * (w * x_out_vector[p].T).flat[0])))))
    return ({'E_out':E_out, 'new_xvals':new_xvals, 'new_yvals':new_yvals, 'output_out':output_out})
```

```python
def gen_n_out(n_out, new_xvals, new_yvals, orig_intercept, orig_grad_line, plot_flag = 'N'):
    output_out = np.zeros(n_out)
    constant_out = np.ones(n_out)

    new_out = genPtsAndPlot(n_out,'N')
    new_xvals = new_out['xvals']
    new_yvals = new_out['yvals']

    for i in range(n_out):
        x = new_xvals[i]
        y = new_yvals[i]
        line_x_val = (y - orig_intercept)/orig_grad_line
        output_out[i] = np.sign(x - line_x_val)


    # Note
    # output variable stores the +/- 1 indication
    output_out[output_out==0]=-1


    # At y = 1, x = ...
    orig_xrnd1 = (1-orig_intercept)/orig_grad_line

    # At y = -1
    orig_xrnd_minus1 = (-1-orig_intercept)/orig_grad_line


    if plot_flag == 'Y':
        # Plot Complete Figure
        plt.figure()
        plt.scatter(new_xvals, new_yvals, alpha=0.5, s = 50, c = output_out)
        plt.plot([orig_xrnd1, orig_xrnd_minus1], [1,-1], color='green')
        plt.show()


    # Calculate Eout

    x_out = pd.Series(np.linspace(0,n_out-1,n_out))
    x_out_vector = np.matrix([constant_out, new_xvals, new_yvals])
    x_out_vector = x_out_vector.T

    return ({'new_xvals':new_xvals, 'new_yvals':new_yvals, 'output_out':output_out})



```
